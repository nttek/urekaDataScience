```python
import pandas as pd
import fasttext
model = fasttext.load_model('lid.176.ftz')
```

```python
reviews = pd.read_csv('Data/reviews.csv')
reviews.head()
```

```python
comments = pd.DataFrame(reviews['comments'])
comments['comments'] = comments['comments'].str.replace(r'\n','')
comments.head()
#type(comments)

df1 = comments.iloc[:20000, :]
df2 = comments.iloc[20000:40000, :]
df3 = comments.iloc[40000:60000, :]
df4 = comments.iloc[60000:, :]

print(df1.head())
print(df1.tail())
print(df2.head())
print(df2.tail())
print(df3.head())
print(df3.tail())
print(df4.head())
print(df4.tail())
```

```python
#Creating a list to save the language label
lang_label = []

for index, row in df1.iterrows(): 
    review = str (row["comments"])   
    t = model.predict(review) 
    lang_label.append(t[0])
    label_acc_column1 = pd.DataFrame(lang_label, columns =['lang_label'])
    
label_acc_column1
```

```python
#Creating a list to save the language label
lang_label = []

for index, row in df2.iterrows(): 
    review = str (row["comments"])   
    t = model.predict(review) 
    lang_label.append(t[0])
    label_acc_column2 = pd.DataFrame(lang_label, columns =['lang_label'])
    
label_acc_column2
```

```python
#Creating a list to save the language label
lang_label = []

for index, row in df3.iterrows(): 
    review = str (row["comments"])   
    t = model.predict(review) 
    lang_label.append(t[0])
    label_acc_column3 = pd.DataFrame(lang_label, columns =['lang_label'])
    
label_acc_column3
```

```python
#Creating a list to save the language label
lang_label = []

for index, row in df4.iterrows(): 
    review = str (row["comments"])   
    t = model.predict(review) 
    lang_label.append(t[0])
    label_acc_column4 = pd.DataFrame(lang_label, columns =['lang_label'])
    
label_acc_column4
```

```python
frames = [label_acc_column1, label_acc_column2, label_acc_column3, label_acc_column4]

result = pd.concat(frames)
```

```python
result.to_csv('Data/result.csv', index=False)
trial = pd.read_csv('Data/result.csv')
trial
```

```python
merged = pd.concat([reviews, trial], axis=1, sort=False)
merged.to_csv('Data/merged.csv', index=False)
merged
```

```python
model = fasttext.train_unsupervised('Data/merged.csv')
```
